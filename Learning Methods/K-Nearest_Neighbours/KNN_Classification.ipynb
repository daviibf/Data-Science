The **kNN (k-Nearest Neighbors)** is a very simple and very known classification method that can handle any number of labels.
Its paradigm consists of labelling the new data by just checking the k-neighbors's labels, where the majority wins the classification. 

![](kNN_fig.png)

In the event of a tie, we 'flip a coin' to choose which of the majority tied labels will be used for classification.

Above, we divide our functions into three: 
> Get the Euclidean Distance: **get_distance()**

> Get the k-Nearest-Neighbours of one point: **get_neighbours()**

> Analyse the k-Nearest-Neighbours for each point and returns the predicted labels: **kNN()**

```python
# First we write a function to get the Euclidean Dist. of 2 points
def get_distance(p1, p2):
    return np.linalg.norm(p2 - p1)

# Than we Get the K-Neighbours of p given all the Train Set
def get_neighbours(x_train, p, k):
    distances = []
    
    for i in range(len(x_train)):
        distances.append(get_distance(x_train[i], p))
        
    min_i = np.argsort(distances)[:k]
    min_d = np.sort(distances)[:k]
    
    return min_i, min_d 

# Finally we can calculate our kNN for each data of the Test Set
def kNN(x_train, y_train, x_test, k, n_labels = 3):
    y_pred = []
    
    # For all the test points, I check for all the train points
    for i in range(len(x_test)):
        knn_idx, _ = get_neighbours(x_train, x_test[i], k)
        knn_idx = [int(q) for q in knn_idx]
        
        kN_labels = y_train[knn_idx]
        kN_labels = [int(p) for p in kN_labels]
        kN_labels = np.array(kN_labels)
        
        # We count the quantity of which label from the k-neighbours
        # in order to know the majority
        possible_labels = np.arange(n_labels)
        
        # Count the occurrence of the labels in the neighbourhood
        count = []
        for j in range(n_labels):
            count.append(len(kN_labels[kN_labels == possible_labels[j]]))
        count = np.array(count)
        
        # Take its indexes into order
        increasing_order = count.argsort()
        
        # Count if we have more than one maximum
        equal_max = 0
        
        # If they are all the same
        if(count[increasing_order[-1]] == count[increasing_order[0]]):
            equal_max = len(count)
            chosen = np.random.randint(0, equal_max) #random intenger in {0, 1, ..., equal_max - 1}
            y_pred.append(increasing_order[chosen])
        
        # If not, count how many are equal
        else:
            while(count[increasing_order[-1]] == count[increasing_order[-2 - equal_max]]):
                equal_max += 1
            
            if(equal_max != 0):
                chosen = np.random.randint(0, equal_max + 1) #random intenger in {0, 1, ..., equal_max}
                y_pred.append(increasing_order[-1 - chosen]) # pick the chosen one among the greatests
            
            # If we don't, just pick the maximum occurrence
            else:
                y_pred.append(increasing_order[-1])
        
    return np.array(y_pred)
```

## 2. k-NN Classification: Szeged-weather dataset

Now, we are going to apply the k-NN Method in a real dataset: **Szeged-weather**.

The dataset has plenty features (as we can see below), but our goal using it is to predict the **Apparent Temperature** using only the **Real Temperature** and the **Humidity**.

```python
data_weather = pd.read_csv("weatherHistory.csv")
data_weather
```

```python
# We want to predict the Apparent Temperature based on the actual Temperature and Humidity
X_data = np.array([data_weather["Temperature (C)"], data_weather["Humidity"]]).T
y_data_ = np.array(data_weather["Apparent Temperature (C)"])
```

## Question 2.1

We can see how the Apparent Temperature behaves in matters of the other two variabels of interest, as exhibited below.

```python
plt.figure(figsize = (10,10))
plt.scatter(X_data[:, 0], X_data[:, 1], c = -y_data_, cmap="Spectral")
plt.xlabel("Temperature (C)", size = 10)
plt.ylabel("Humidity", size = 10)
plt.show()
```

As expected, the Apparent Temperature somehow follows the True Temperature, but also the Humidity has some influence in its value, making it not so trivial to predict.

In order to use the KNN, we can separate the data in 5 groups based on the Apparent Temperature.

```python
y_data = []
clusters = 5

step = (y_data_.max() - y_data_.min()) / clusters

for j in range(len(y_data_)):
    for i in range(clusters):
        if(y_data_[j] <= (y_data_.min() + step * (i + 1))):
            y_data.append(i)
            break
            
y_data = np.array(y_data)
```

After separating this data in 5 Clusters, we have our new representation based on the Apparent Temperature:

## Question 2.2

We will now pick the 2000 first samples of the dataset and separate the data into 5 folds randomly.

```python
n_data = 2000

# Pick the first 2000 data
X = X_data[:n_data]
y = y_data[:n_data]

# Separate randomly in 5 folds
n_folds = 5
seq = np.arange(len(X))
np.random.shuffle(seq)
X = X[seq]
y = y[seq]

X_folds = []
y_folds = []
for i in range(n_folds):
    X_folds.append(X[int(i * n_data/n_folds):int((i + 1) * n_data/n_folds)])
    y_folds.append(y[int(i * n_data/n_folds):int((i + 1) * n_data/n_folds)])

X_folds = np.array(X_folds)
y_folds = np.array(y_folds)
```

We will choose a fold among them to be used as a **Validation Fold**, while the remaining ones will be together the **Training Data**. Each fold will be the Validation Fold once and in the end we will have the errors for each scenario.

This process will repeat for each one of the K's we are testing, so in the end we have all **Means** and **Standard Deviations** of the errors for each K:

$$k \in \big{\{}1, 2, ..., 10\big{\}}$$

```python
# Each K to be tested
K = np.arange(1, 11)
```

```python
errors = []

# For each K run the KNN
for k in K:
    error_aux = []
    
    # For each fold, run the method and compute the errors
    for j in range(n_folds):
        X_train = np.concatenate(X_folds[np.arange(n_folds) != j])
        y_train = np.concatenate(y_folds[np.arange(n_folds) != j])

        X_test = X_folds[j]
        y_test = y_folds[j]
        
        y_pred = kNN(X_train, y_train, X_test, k, clusters)

        err = 0
        for j in range(len(y_pred)):
            if (y_pred[j] != y_test[j]):
                err += 1
                
        error_aux.append(err/len(y_pred))
    
    errors.append(error_aux)
    
```

```python
# Take the mean and standard deviation to plot
mean_errors = np.mean(errors, axis = 1)
sd_errors = np.std(errors, axis = 1)
```

Now that the errors have been acquired, we can see its behavior with an **Error Bar Representation** for each K value used.

```python
plt.figure(figsize = (15,5))
plt.title('Error Bar Representation for each K', fontsize=25)
plt.errorbar(K, mean_errors, fmt = 'o', yerr = sd_errors)
plt.xlabel("K nearest neighbours", size=25)
plt.ylabel("KNN's missclassification", size=15)
plt.show()
```

The confusion matrix can be very helpful for a visual analyse. It consists of the number of correct and incorrect predictions which are summarized with count values and broken down by each class.

Thus, we can analyze which labels our model is most confusing and which labels are impersonating others more often.

```python
def confusion_matrix(y,y_pred,classes,l=10,title='Confusion matrix',normalize=False, colored = True):
    
    # below we count the occurrences of each
    # label and how many predictions were correct and how many were not.
    
    # The Confusion Matrix
    cm = []
    for i in range(l):
        
        # The rows to be updated
        row = np.zeros(l)
        for j in range(len(y)):
            if (y[j] == y_pred[j] == i+1):
                row[i] += 1
            else:
                if(y[j] == i+1):
                    row[y_pred[j]] += 1
                    
        # Each row is appended into the matrix
        cm.append(row)
    cm = np.array(cm)
    
    if colored:
        # Below is a code that I found for coloring the matrix:
        fig, ax = plt.subplots()
        im = ax.imshow(cm, cmap=plt.cm.Blues)
        ax.figure.colorbar(im, ax=ax)

        ax.set(xticks=np.arange(cm.shape[1]),
               yticks=np.arange(cm.shape[0]), 
               xticklabels = classes, 
               yticklabels = classes,
               title=title,
               ylabel='Original labels',
               xlabel='Predicted labels')

        plt.setp(ax.get_xticklabels(), ha="right")

        fmt = '.2f' if normalize else '.0f'
        thresh = cm.max() / 2.
        for i in range(cm.shape[0]):
            for j in range(cm.shape[1]):
                ax.text(j, i, format(cm[i, j], fmt),
                        ha="center", va="center",
                        color="white" if cm[i, j] > thresh else "black")
        fig.tight_layout()

        return cm, ax
    else:
        return cm
```
