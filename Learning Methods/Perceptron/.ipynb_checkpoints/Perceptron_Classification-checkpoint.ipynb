```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_gaussian_quantiles
```

# 1. Synthetic Data

#### Parameters for the datasets:

```python
#Â Variance of the data
sigma_2 = [0.05, 0.25, 0.50, 0.75]
rate = [0.1, 0.25, 0.5, 0.75]

# Mean vectors
mu_1 = [-1, 0]
mu_2 = [1, 0]
```

## Question 1.1:
### Analyze the Data behavior based on the variance:

There are several hypothesis classes which belongs to Linear Predictors: Linear Regression, Logistic Regression, Halfspaces...

In our case, we are dealing with Halfspaces and how good the Perceptron Algorithm can be in these situations.

The halfspaces are designed for binary classification problems, with $X \in R^{d}$ and $Y \in \{0, 1\}$. Our parameters are the coefficients of the linear classification we wanna find plus a Bias. Given a matrix X of data, our perceptron returns the $w \in R^d$ and the $b \in R$. 

The goal is to find the vector $w$ whose hyperplane is perpendicular to it and separates the data into two labels.

Also, the perceptron Algorithm can be very useful for the 'realizable cases', also refered to as the 'separable case', since it is possible to divide all the 1's from 0's with an hyperplane. Later on we will see that as the data labels start to 'mix'(either by changing the variance or by inserting a probability factor), the more ineffective the Perceptron Algorithm becomes.

The training of the model is done by the following system of equations:

**-If the model find 1 where the data is actually 0, it does:**

$w_{i+1} = w(n)-\eta(n)x(n)\; \; if \; \; w^{T}(n)x(n)>0$

**-If the model find 0 where the data is actually 1, it does:**

$w_{i+1} = w(n)+\eta(n)x(n)\; \; if \; \; w^{T}(n)x(n)\leq0$

#### Perceptron function:

```python
def perceptron(X, y, learning_rate = 0.5):
    # Insert a column of ones for the bias value
    X = np.insert(X, 0, np.ones(len(X)), axis = 1)
    
    # Initialize w
    w = np.zeros(len(X[0]))
    
    # Training
    for j in range(len(X)):
        # If the classification is wrong:        
        if(np.dot(w, X[j]) > 0 and y[j] == 0):
            w = w - learning_rate * X[j]
            
        if(np.dot(w, X[j]) <= 0 and y[j] == 1):
            w = w + learning_rate * X[j]
    
    # Testing
    e = 0
    
    for j in range(len(X)):
        # If the classification is wrong:
        if(np.dot(w, X[j]) > 0 and y[j] == 0):
            e += 1
            
        if(np.dot(w, X[j]) <= 0 and y[j] == 1):
            e += 1
    
    # Returns the weights and the fraction missclassified points
    return(X, w, e/len(X))
```

In our approach, we incorporate bias as an element of w, and add a column of 1's in our matrix X. Thus, our line equation becomes:

$$w_0 + x_0 \cdot w_1 + x_2 \cdot w_2 = 0$$
$$ x_2 = -\frac{1}{w_2}(w_0 + w_1 \cdot x_1) $$

$$Now, \; X,w \in R^{d+1}$$

Below, we shuffle our data X in order to garantee that our perceptron will learn randomly which is the better hyperplane, instead of learning all the 0's, and just than checking all the 1's. Doing this, we avoid possible bias of choosing a side more than another.

### Analyse the perceptron with a fixed Learning Rate and differents Variances.

```python
titles = ["Variance = 0.05", "Variance = 0.25", "Variance = 0.50", "Variance = 0.75"]

plt.figure(figsize = (20, 10))

for i in range(4):
    #Â Creates half of the data with mean mu_1 and the other half mu_2
    x1, _ = make_gaussian_quantiles(mu_1, sigma_2[i] * np.identity(2), 100, 2)
    x2, _ = make_gaussian_quantiles(mu_2, sigma_2[i] * np.identity(2), 100, 2)
    X = np.append(x1, x2, axis = 0)
    
    # Shuffle the data to avoid biased weights
    seq = np.arange(len(X))
    np.random.shuffle(seq)
    X = X[seq]
    
    # Associate the y's with the x's
    y = np.append(np.zeros(100), np.ones(100))
    y = y[seq]
    
    # Run the perceptron
    XX, w, e = perceptron(X, y, learning_rate=0.1)
    print(w)
    
    # Hiperplane function from x2 = f(x1)
    x1_hp = np.linspace(min(XX[:,1])-0.5, max(XX[:,1])+0.5, 200)
    x2_hp = -(w[0]+ w[1]*x1_hp)/(w[2])
    
    plt.subplot(2, 2, i + 1)
    plt.title(titles[i], size = 15)

    plt.plot(XX[y == 0, 1], XX[y == 0, 2], 'ro', label = 'y = 0')
    plt.plot(XX[y == 1, 1], XX[y == 1, 2], 'bo', label = 'y = 1')
    plt.plot(x1_hp, x2_hp)
    
    plt.ylim(min(XX[:,2])-0.5 ,max(XX[:,2])+0.5)
    plt.xlim(min(XX[:,1])-0.5 ,max(XX[:,1])+0.5)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()

plt.show()
```

### Analyse the perceptron behavior for the Separable Case with differents Learning Rates.

```python
titles = ["Learning Rate = 0.1", "Learning Rate = 0.25", "Learning Rate = 0.50", "Learning Rate = 0.75"]

#Â Creates half of the data with mean mu_1 and the other half mu_2
x1, _ = make_gaussian_quantiles(mu_1, sigma_2[2] * np.identity(2), 100, 2)
x2, _ = make_gaussian_quantiles(mu_2, sigma_2[2] * np.identity(2), 100, 2)
X = np.append(x1, x2, axis = 0)

# Shuffle the data to avoid biased weights
seq = np.arange(len(X))
np.random.shuffle(seq)
X = X[seq]
    
# Associate the y's with the x's
y = np.append(np.zeros(100), np.ones(100))
y = y[seq]


plt.figure(figsize = (20, 10))

for i in range(4):    
    # Run the perceptron
    XX, w, e = perceptron(X, y, learning_rate=rate[i])
    
    # Hiperplane function from x2 = f(x1)
    x1_hp = np.linspace(min(XX[:,1])-0.5, max(XX[:,1])+0.5, 200)
    x2_hp = -(w[0]+ w[1]*x1_hp)/(w[2])
    
    plt.subplot(2, 2, i + 1)
    plt.title(titles[i], size = 15)

    plt.plot(XX[y == 0, 1], XX[y == 0, 2], 'ro', label = 'y = 0')
    plt.plot(XX[y == 1, 1], XX[y == 1, 2], 'bo', label = 'y = 1')
    plt.plot(x1_hp, x2_hp)
    
    plt.ylim(min(XX[:,2])-0.5 ,max(XX[:,2])+0.5)
    plt.xlim(min(XX[:,1])-0.5 ,max(XX[:,1])+0.5)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()

plt.show()
```

#### For each Variance, run over 50 randomly generated datasets.

```python
# Keep the error ratios
errors = []

for i in range(len(sigma_2)):
    aux = []
    
    for k in range(50):
        #Â Creates half of the data with mean mu_1 and the other half mu_2
        x1, _ = make_gaussian_quantiles(mu_1, sigma_2[i] * np.identity(2), 100, 2)
        x2, _ = make_gaussian_quantiles(mu_2, sigma_2[i] * np.identity(2), 100, 2)
        X = np.append(x1, x2, axis = 0)
    
        # Shuffle the data to avoid biased weights
        seq = np.arange(len(X))
        np.random.shuffle(seq)
        X = X[seq]
    
        # Associate the y's with the x's
        y = np.append(np.zeros(100), np.ones(100))
        y = y[seq]
        
        # Run the perceptron
        XX, w, e = perceptron(X, y)
        
        # Keep the errors for each dataset
        aux.append(e)
        
    errors.append(aux)
    
# Take the mean and standard deviation for the errors of each sigma_2
mean_errors = np.mean(errors, axis = 1)
sd_errors = np.std(errors, axis = 1)
```

```python
plt.figure(figsize = (15,5))

plt.title('Error Bar Representation for each Variance', fontsize=25)
plt.errorbar(sigma_2, mean_errors, fmt = 'o', yerr = sd_errors)
plt.xlabel("Variance of the Dataset", fontsize=25)
plt.ylabel("Perceptron's missclassification", fontsize=15)
plt.show()
```

## Comment:

So, as expected, we can see that the perceptron is capable of creating a good model while the data variances support them in a separable case.

As the variation increases between the two types, the greater the standard deviation and the less accurate the hyperplane becomes.

The Learning Rate, as well, can help us achieving a good model. With a lower Learning Rate, it takes more time to converge but we usually get a more accurate model. On the other hand, a high Learning Rate can converge easily, but with a not so goog fit. 

## Question 1.2:

### Analyze the Data behavior based on the probability:

```python
#Â A new dataset with sigma_2 = 0.15 is created
x1, _ = make_gaussian_quantiles(mu_1, 0.15 * np.identity(2), 100, 2)
x2, _ = make_gaussian_quantiles(mu_2, 0.15 * np.identity(2), 100, 2)
X = np.append(x1, x2, axis = 0)

# Shuffle the data to avoid biased weights
seq = np.arange(len(X))
np.random.shuffle(seq)
X = X[seq]

# Probabilities of flipping y
prob = [0, 0.05, 0.10, 0.20]
```

```python
titles = ["p = 0.00", "p = 0.05", "p = 0.10", "p = 0.20"]

plt.figure(figsize = (20, 10))

for i in range(4):
    # Create y as it usually is
    y = np.append(np.zeros(100, int), np.ones(100, int))
    y = y[seq]
    
    for k in range(len(X)):
        # With probability p:
        if(np.random.rand() <= prob[i]):
            # Flip the value with a xor operation
            y[k] = y[k] ^ 1
    
    # Run the perceptron
    XX_prob, w, e = perceptron(X, y, learning_rate=0.1)
    
    # Hiperplane function from x2 = f(x1)
    x1_hp_prob = np.linspace(min(XX_prob[:,1])-0.5, max(XX_prob[:,1])+0.5, 200)
    x2_hp_prob = -(w[0]+ w[1]*x1_hp_prob)/(w[2])
    
    plt.subplot(2, 2, i + 1)
    plt.title(titles[i], size = 15)

    plt.plot(XX_prob[y == 0, 1], XX_prob[y == 0, 2], 'ro', label = 'y = 0')
    plt.plot(XX_prob[y == 1, 1], XX_prob[y == 1, 2], 'bo', label = 'y = 1')
    plt.plot(x1_hp_prob, x2_hp_prob)
    
    plt.ylim(min(XX_prob[:,2])-0.5 ,max(XX_prob[:,2])+0.5)
    plt.xlim(min(XX_prob[:,1])-0.5 ,max(XX_prob[:,1])+0.5)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    
plt.show()
```

Repeat the previous experiment:

```python
# Keep the missclassifications for each p
errors_prob = []

for p in prob:
    aux_prob = []
    
    for k in range(50):
        # Initialize y as it normally is
        y = np.append(np.zeros(100, int), np.ones(100, int))
        y = y[seq]
        
        for i in range(len(X)):
            # With probability p:
            if(np.random.rand() <= p):
                # Flip the value with a xor operation
                y[i] = y[i] ^ 1

        # Run the perceptron
        XX, w, e = perceptron(X, y)
        
        aux_prob.append(e)
    
    errors_prob.append(aux_prob)

# Take the mean and standard deviation for the errors of p
mean_errors_p = np.mean(errors_prob, axis = 1)
sd_errors_p = np.std(errors_prob, axis = 1)
```

```python
plt.figure(figsize = (15,5))
plt.title('Error Bar Representation for each Probability', fontsize=25)
plt.errorbar(prob, mean_errors_p, fmt = 'o', yerr = sd_errors_p)
plt.xlabel("Probability of Flipping", size=25)
plt.ylabel("Perceptron's missclassification", size=15)
plt.show()
```

### Comment:

When dealing with the probabilistic factor, we dramatically increase the chance of having outliers in our data. A quick analysis in the Scatter Plot of the data for each value of p, we can see that the flipped labels do not follow any pattern in relation to the hyperplane that we must plot, consequently, our standard deviation goes up a lot compared to the situation simulated in the first question.

Below we have the two Erro Bar Plors, side by side, to a better comparison of how the error changes from one to another. 

### Comparing the Error's Standard Deviation for each case.

```python
plt.figure(figsize = (15,15))

plt.subplot(2, 1, 1)
plt.title('Graph to compare the missclassification of both cases', fontsize=25)

plt.errorbar(prob, mean_errors_p, fmt = 'o', yerr = sd_errors_p)
plt.xlabel("Probability of Flipping", size=15)
plt.ylabel("missclassification", size=15)

plt.subplot(2, 1, 2)
plt.errorbar(sigma_2, mean_errors, fmt = 'o', yerr = sd_errors)
plt.xlabel("Variance of the Dataset", fontsize=15)
plt.ylabel("missclassification", fontsize=15)
plt.show()
```
